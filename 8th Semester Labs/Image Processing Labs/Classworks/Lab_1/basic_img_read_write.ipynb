```python
import numpy as np
import matplotlib.pyplot as plt

width, height = 100, 100

image_array = np.zeros((height, width), dtype=np.uint8)

for i in range(height):
    for j in range(width):
        image_array[i, j] = (i + j) % 256  # simple gradient effect

# Display the image using matplotlib
plt.imshow(image_array, cmap='gray')
plt.title('Black & White Image using Numpy')
plt.axis('off')  # Turn off the axes for better visualization
plt.show()
```

```python
# Save the pixel values (2D array) to a text file
import math 
file_path = 'image_pixel_values.txt'
np.savetxt(file_path, image_array, fmt='%d')

# Read the pixel values from the text file
loaded_image_array = np.loadtxt(file_path, dtype=np.uint8)

for i in range(len(loaded_image_array)):    
    for j in range(len(loaded_image_array[0])):
        loaded_image_array[i, j] = abs(j-i)%256

# Display the image loaded from the text file to ensure it's the same
plt.imshow(loaded_image_array, cmap='gray')
plt.title('Reconstructed Image from Text File')
plt.axis('off')
plt.show()

file_path  # Return the file path so the user can download and explore the pixel values
```
