```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python


# Read the grayscale image
image = cv2.imread('../images/apple.jpg', cv2.IMREAD_GRAYSCALE)

# Apply a threshold
threshold_value = 210  # Can vary depending on the application
_, binary_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)

# Display original and binary images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Grayscale Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Binary Image after Thresholding')
plt.imshow(binary_image, cmap='gray')
plt.axis('off')

plt.show()
```

```python
from PIL import Image

def load_image(image, w, h):

    # Convert the binary data to an image
    image = Image.new('L', (w, h))  # 'L' mode for grayscale, size 5x5
    pixels = image.load()

    # Fill the image with the pixel values from the binary image array
    for i in range(image.size[0]):
        for j in range(image.size[1]):
            pixels[i, j] = binary_image[j][i]  # Note: (x, y) is (i, j)
            
    return image
```

```python
# Sample grayscale image represented as a 2D list (5x5 for simplicity)
grayscale_image = [
    [34, 120, 200, 45, 90],
    [12, 180, 255, 80, 150],
    [100, 60, 75, 190, 230],
    [0, 50, 170, 210, 140],
    [255, 220, 30, 160, 110]
]

image = load_image(grayscale_image, 5, 5)
image.show()
```

```python
from PIL import Image

# Sample grayscale image represented as a 2D list (5x5 for simplicity)
grayscale_image = [
    [34, 120, 200, 45, 90],
    [12, 180, 255, 80, 150],
    [100, 60, 75, 190, 230],
    [0, 50, 170, 210, 140],
    [255, 220, 30, 160, 110]
]

# Convert the binary data to an image
image = Image.new('L', (5, 5))  # 'L' mode for grayscale, size 5x5
pixels = image.load()

# Define the threshold value
threshold_value = 127

# Apply thresholding
binary_image = []
for row in grayscale_image:
    binary_row = []
    for pixel in row:
        # If pixel value is greater than threshold, set to 255 (white)
        # Otherwise, set to 0 (black)
        if pixel > threshold_value:
            binary_row.append(255)
        else:
            binary_row.append(0)
    binary_image.append(binary_row)

# Print the resulting binary image
print("Binary Image:")
for row in binary_image:
    print(row)

```
