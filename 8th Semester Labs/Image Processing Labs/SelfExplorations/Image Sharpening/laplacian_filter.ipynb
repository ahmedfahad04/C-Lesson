```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the image (grayscale for simplicity)
image = cv2.imread('bird.jpg', cv2.IMREAD_GRAYSCALE)

# Step 2: Define the Laplacian kernel (3x3 kernel)
laplacian_kernel = np.array([[0, -1,  0],
                             [-1,  4, -1],
                             [0, -1,  0]])

# Step 3: Apply the Laplacian filter (convolution)
filtered_image = cv2.filter2D(image, -1, laplacian_kernel)

# Step 4: Sharpen the image by adding the Laplacian result back to the original
sharpened_image = cv2.addWeighted(image, 1.0, filtered_image, 0.5, 0)

# Step 5: Display the results
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 3, 2)
plt.title('Laplacian Filtered')
plt.imshow(filtered_image, cmap='gray')

plt.subplot(1, 3, 3)
plt.title('Sharpened Image')
plt.imshow(sharpened_image, cmap='gray')

plt.show()
```
