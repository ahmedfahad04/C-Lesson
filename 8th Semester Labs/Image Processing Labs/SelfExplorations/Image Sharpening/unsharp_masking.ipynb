```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the image (grayscale for simplicity)
image = cv2.imread('../../images/moon.jpg', cv2.IMREAD_GRAYSCALE)

# Step 2: Blur the original image using a Gaussian filter
blurred_image = cv2.GaussianBlur(image, (9, 9), 10)  # kernel size and sigma can be adjusted

# Step 3: Create the mask by subtracting the blurred image from the original
mask = cv2.subtract(image, blurred_image)

# Step 4: Apply Unsharp Masking - Add the mask back to the original image
unsharp_image = cv2.addWeighted(image, 1.0, mask, 2, 0)  # Change weights to control sharpening

# Step 5: Apply High-Boost Filtering
boost_factor = 1.2  # High boost factor (can adjust for stronger effect)
high_boost_image = cv2.addWeighted(image, boost_factor, mask, 1.0, 0)

# Step 6: Display results
plt.figure(figsize=(20, 15))

plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(2, 2, 2)
plt.title('Blurred Image')
plt.imshow(blurred_image, cmap='gray')

plt.subplot(2, 2, 3)
plt.title('Unsharp Masking')
plt.imshow(unsharp_image, cmap='gray')

plt.subplot(2, 2, 4)
plt.title('High-Boost Filtered')
plt.imshow(high_boost_image, cmap='gray')

plt.show()
```

```python
# Step 1: Load the image (grayscale for simplicity)
image = cv2.imread('../../images/moon.jpg', cv2.IMREAD_GRAYSCALE)  # cv2.IMREAD_GRAYSCALE for grayscale

# Step 2: Blur the original image using a Gaussian filter
blurred_image = cv2.GaussianBlur(image, (9, 9), 10)  # (ksize=(9, 9), sigmaX=10)

# Step 3: Create the mask by subtracting the blurred image from the original
mask = cv2.subtract(image, blurred_image)  # (src1=image, src2=blurred_image)

# Step 4: Convert the mask to a binary image using thresholding
_, binary_mask = cv2.threshold(mask, 150, 255, cv2.THRESH_BINARY)  # (src=mask, thresh=30, maxval=255, type=cv2.THRESH_BINARY)

# Step 5: Apply Unsharp Masking - Add the mask back to the original image
unsharp_image = cv2.addWeighted(image, 1.0, mask, 1.5, 1.5)  # (src1=image, alpha=1.0, src2=mask, beta=1.0, gamma=0)

# Step 6: Display results
plt.figure(figsize=(20, 12))  # (figsize=(15, 8))

plt.subplot(2, 2, 1)  # (nrows=2, ncols=3, index=1)
plt.title('Original Image')  # (label='Original Image')
plt.imshow(image, cmap='gray')  # (X=image, cmap='gray')

plt.subplot(2, 2, 2)  # (nrows=2, ncols=3, index=2)
plt.title('Blurred Image')  # (label='Blurred Image')
plt.imshow(blurred_image, cmap='gray')  # (X=blurred_image, cmap='gray')

plt.subplot(2, 2, 3)  # (nrows=2, ncols=3, index=3)
plt.title('Mask')  # (label='Mask')
plt.imshow(mask, cmap='gray')  # (X=mask, cmap='gray')

# plt.subplot(2, 3, 4)  # (nrows=2, ncols=3, index=4)
# plt.title('Binary Mask')  # (label='Binary Mask')
# plt.imshow(binary_mask, cmap='gray')  # (X=binary_mask, cmap='gray')

plt.subplot(2, 2, 4)  # (nrows=2, ncols=3, index=5)
plt.title('Unsharp Masking')  # (label='Unsharp Masking')
plt.imshow(unsharp_image, cmap='gray')  # (X=unsharp_image, cmap='gray')

plt.show()  # Display the plots
```
